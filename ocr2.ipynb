```python
import cv2
import matplotlib.pyplot as plt 
img1 = cv2.imread(r"C:\Users\Owner\Pictures\pytest\test_1.png") #이미지 읽기 
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY) #회색조 변환
blurred = cv2.GaussianBlur(gray, (5, 5), 0) #가우시안 블러 적용 ( 노이즈 제거) 
_, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV) #이진화 적용

#이진화된 이미지 표시
plt.xticks([])
plt.yticks([])
plt.imshow(binary, cmap='gray')
```
